# Importing Libraries

```python
import pandas as pd
import numpy as np
from sklearn import linear_model
```

# DataSet Loading

```python
df = pd.read_excel('Dataset__ai_3.xlsx')
df.head(7)
```

# Data Pre-Processing

```python

column_mapping = {
                  'Unnamed: 1': 'I',
                  'Unnamed: 2': 'TON',
                  'Unnamed: 3': 'SR',
                 'Unnamed: 4': 'WLT',
                 'Unnamed: 5': 'MRR'}

# Rename columns using the rename method
df.rename(columns=column_mapping, inplace=True)
```

```python
df.dropna(inplace=True)
```

```python
df
```

```python
df
```

```python
df.drop(columns=['Data set'],inplace=True)
df = df.reset_index(drop=True)
```

```python
df.head()
```

```python
inputs=df[['I','TON']]
```

```python
inputs.head()
```

```python
outputs=df[['SR','WLT','MRR']]
```

```python
outputs.head()
```

# Create Liner Regression Model

```python
model= linear_model.LinearRegression()
```

# Fit Liner Regression Model

```python
model.fit(inputs,outputs)
```

# Predictions

```python
pre=model.predict([[16,400]])
```

```python
pre
```

# Saving Model

```python
import joblib

joblib.dump(model, 'prediction_model_2.pkl')  
```

```python
# Scientific notation outputs
scientific_outputs = pre[0]

# Convert to real numbers
real_numbers = [float(output) for output in scientific_outputs]

print(real_numbers)
```

```python
real_numbers
```

```python

rounded_values = []
for real_number in real_numbers:
    rounded_values.append(round(real_number, 4))
print(rounded_values) 
```
